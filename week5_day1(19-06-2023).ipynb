### Matplotlib:
        - To display the data Visualization library.
        - Matplotlib
           ->Mat -> mathematical values
           ->plot -> chart
           ->lib -> Library
        - axes:
           -> x-axis value
           -> y-axis value
        - styles:
                markers:o-circle,s-square,D-diamand,p-pentagon,h-hexagon,.->points,+,*,|,-...etc
         -> first we need to import the matplotlib->pyplot
         - syntax: from matplotlib import pyplot as plt.
       - color:
              ->edgecolor -> mec->markeredgecolor
              ->facecolor -> mfc->markerfacecolor
      - linestyle:
              -> ls => solid,dashed,dotted,dashdot,none...

```python
from matplotlib import pyplot as plt
x = [10,20,30,40]
y = [4,6,8,10]
plt.plot(x,y,marker='h',mec='black',mfc='red',ls='dashed',lw='7',ms=10,color='green')
plt.title('Example on markers')
plt.show()
```

```python
import numpy as np
x = np.arange(0,10,2)
y1 =x**2
y2 =x**3
plt.plot(x,y1,label='Squares')
plt.plot(x,y2,label='Cubes')
plt.legend()
plt.xlabel('x-axis values 0 to 10')
plt.ylabel('Y-axis values')
plt.title('Multiple graphs')
plt.show()
```

```python
# Grid graphs...
x=[1,2,3,4]
y=[20,30,10,40]
plt.plot(x,y)
plt.grid(ls='dashed',color='r')
plt.show()
```

### Types of plottings

#### 1. line plot:
- def : A line plot is a graph that shows frequencey of data along with a numbers.
- It is quick , simple way to organize the data..

```python
x=[10,12,14,16]
y=[5,6,7,8]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph')
plt.show()
```

### Bar graph:
- You can use bar() function to draw a bar graphs.
- plt.bar()

```python
year=[2021,2022,2023,2024]
goldprice=[1000,2000,4000,5000]
plt.bar(year,goldprice)
plt.title('bar graph')
plt.show()
```

```python
# Horizontal bargraph
plt.barh(year,goldprice,color=['r','g','w','y'],edgecolor='black')
plt.xlabel('years')
plt.ylabel('goldprice')
plt.show()
```

### scatter plot:
- A scatter plot is a diagram where each value in the dataset represented by a dot.
- plt.scatter()

```python
x=[2,4,6,8]
y=[1,2,3,4]
plt.scatter(x,y,color='pink')
plt.show()
```
