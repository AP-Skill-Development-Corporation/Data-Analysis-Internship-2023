### Data Analysis with pandas:
- merging:

```python
d1={'Name':['nani','ganesh','krishna','Aadyan'],
     'Fruit':['Orange','apple','graps','mango'],
     'color':['black','blue','white','pink']
   }
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame(d1)
df1
```

```python
d2={'Name':['aadyan','krishna','nethra','mani'],
    'Fruit':['orange','banana','apple','mango'],
     'age':[23,45,67,89]}
df2= pd.DataFrame(d2)
df2
```

```python
df1.append(df2)
```

```python
df1
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df3=pd.merge(df1,df2)
df3
```
